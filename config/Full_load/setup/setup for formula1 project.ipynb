### creating external location for formula1 project


```python
%fs ls 'abfss://formula1@azureprostorageaccext.dfs.core.windows.net/landing'
```

```python
%sql
create external location if not exists my_external_location
URL 'abfss://formula1@azureprostorageaccext.dfs.core.windows.net/'
with (storage credential azureprojectextsc)
comment'external location for formula containor';
```

creating catalog for formula1 project

```python
%sql
create catalog if not exists formula1
 managed location 'abfss://formula1@azureprostorageaccext.dfs.core.windows.net/';
```

```python
%sql
SHOW CATALOGS
```

crating schema for formula1

```python
%sql
create schema if not exists formula1.landing;
create schema if not exists formula1.bronze
  managed location 'abfss://formula1@azureprostorageaccext.dfs.core.windows.net/bronze';
  create schema if not exists formula1.silver
  managed location 'abfss://formula1@azureprostorageaccext.dfs.core.windows.net/silver';
  create schema if not exists formula1.gold
  managed location 'abfss://formula1@azureprostorageaccext.dfs.core.windows.net/gold';
```

```python
%sql use catalog formula1;
```

```python
%sql
show schemas;
```
