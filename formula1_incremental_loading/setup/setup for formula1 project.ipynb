### creating external location for formula1 project


```python
%fs ls 'abfss://formula1-incr@azureprostorageaccext.dfs.core.windows.net/landing'
```

```python
%sql
create external location if not exists my_external_location_incr
URL 'abfss://formula1-incr@azureprostorageaccext.dfs.core.windows.net/'
with (storage credential azureprojectextsc)
comment'external location for formula1 incremental containor';
```

creating catalog for formula1 project

```python
%sql
create catalog if not exists formula1_incr
 managed location 'abfss://formula1-incr@azureprostorageaccext.dfs.core.windows.net/';
```

```python
%sql
SHOW CATALOGS
```

crating schema for formula1

```python
%sql
create schema if not exists formula1_incr.landing;
create schema if not exists formula1_incr.bronze
  managed location 'abfss://formula1-incr@azureprostorageaccext.dfs.core.windows.net/bronze';
  create schema if not exists formula1_incr.silver
  managed location 'abfss://formula1-incr@azureprostorageaccext.dfs.core.windows.net/silver';
  create schema if not exists formula1_incr.gold
  managed location 'abfss://formula1-incr@azureprostorageaccext.dfs.core.windows.net/gold';
```
