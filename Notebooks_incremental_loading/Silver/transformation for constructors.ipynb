```python
bronze_table = f"{catalog_name}.{bronze_schema}.constructors"
silver_table = f"{catalog_name}.{silver_schema}.constructors"
```

```python
from pyspark.sql import functions as F
```

```python
#read data from brozne_table
df_constructors = (spark.read.table(bronze_table).filter(F.col("batch_id") == v_batch_id))
```

```python
#drop url col
df_constructors_dropped = df_constructors.drop('url')
```

```python
#rename columns
df_constructors_renamed = df_constructors_dropped.withColumnsRenamed({
    "constructorId": "constructor_id",
    "name": "constructor_name"
})
```

```python
#drop duplicates
df_constructors_distinct = df_constructors_renamed.dropDuplicates(['constructor_id'])
```

```python
#final datafarame
df_constructors_final = df_constructors_distinct.withColumn('nationality' , F.initcap(F.col('nationality')))
```

```python
write_to_silver(input_df=df_constructors_final, target_table=silver_table,
                merge_condition="t.constructor_id = s.constructor_id",
                columns_to_update=['constructor_name',
                                   'nationality',
                                   'ingestion_timestamp',
                                   'source_file',
                                   'batch_id'])
```
