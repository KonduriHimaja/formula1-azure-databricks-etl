```python
source_file = f"{loading_folder_path}/constructors.json"
table_name = f"{catalog_name}.{bronze_schema}.constructors"
```

```python
#define schema
constructor_schema = """constructorId STRING,
                        name STRING,
                        nationality STRING,
                        url STRING"""
```

```python
#read data from the file
constructor_df = (
    spark.read
          .format('json')
          .schema(constructor_schema)
          .option('mode', 'failfast')
          .load(source_file)
)
```

```python
final_df_constructors = add_ingestion_metadata(constructor_df)
```

```python
(final_df_constructors
     .write
     .format("delta")
     .mode("overwrite")
     .saveAsTable(table_name))
```
