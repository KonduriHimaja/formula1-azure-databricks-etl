```python
source_file = f"{loading_folder_path}/circuits.csv"
table_name = f"{catalog_name}.{bronze_schema}.circuits"
source_file1 = f"{loading_folder_path}/races.csv"
table_name1 = f"{catalog_name}.{bronze_schema}.races"
```

```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType

circuits_schema = StructType([
    StructField('circuitId', StringType()),
    StructField('url', StringType()),
    StructField('circuitName', StringType()),
    StructField('lat', DoubleType()),
    StructField('long', DoubleType()),
    StructField('locality', StringType()),
    StructField('country', StringType())
])
```

```python
df_circuits = (
    spark.read
         .format('csv')
         .option('header', 'true')
         .option('mode', 'failfast')
         .schema(circuits_schema)
         .load(source_file))
```

adding metadata column

```python
final_df_circuits = add_ingestion_metadata(df_circuits)
```

write data to detla table in bronze layer


```python
(
    final_df_circuits
      .write
      .format('delta')
      .mode('overwrite')
      .saveAsTable(table_name)
)
```

```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType, DateType

races_schema = StructType([
    StructField('season', IntegerType()),
    StructField('round', IntegerType()),
    StructField('url', StringType()),
    StructField('raceName', StringType()),
    StructField('date', DateType()),
    StructField('circuitId', StringType())
])
```

```python
df_races = (
    spark.read
         .format('csv')
         .option('header', 'true')
         .option('mode', 'failfast')
         .schema(races_schema)
         .load(source_file1))
```

```python
final_df_races = add_ingestion_metadata(df_races)
```

```python
(
    final_df_races
      .write
      .format('delta')
      .mode('overwrite')
      .saveAsTable(table_name1)
)
```
